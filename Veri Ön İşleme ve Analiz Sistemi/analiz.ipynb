```python
from kutuphane import DataPreprocessor
import pandas as pd
import matplotlib.pyplot as plt
```

## Veri Setinin Yüklenmesi

Bu adımda `netflix_titles.csv` dosyası `DataPreprocessor` sınıfı kullanılarak programa yüklenir.v

```python
verisetim = DataPreprocessor('veri/netflix_titles.csv')
```

## Veri Setinin Genel Özeti

Bu işlem ile:
- İlk 5 satır görüntülenir.
- Sütunların veri tipleri incelenir.
- Sayısal sütunların istatistiksel özeti gösterilir.

```python
verisetim.veri_ozeti()
```

## Eksik Verilerin Kontrolü

Her sütunda bulunan eksik veri sayısı görüntülenir.

Netflix veri setinde özellikle `director`, `cast` ve `country` sütunlarında eksik değerler bulunmaktadır.

```python
verisetim.eksik_veri_kontrolu()
```

## Eksik Verilerin Doldurulması

Eksik değerler uygun yöntemlerle doldurulur:
- Sayısal sütunlarda ortalama veya medyan
- Metinsel sütunlarda en sık görülen değer 

```python
verisetim.eksikleri_doldur(strateji='mean')
```

## Gereksiz Sütunun Silinmesi

`show_id` sütunu her satır için benzersiz kimlik bilgisidir.
Analiz açısından anlamlı olmadığı için veri setinden çıkarılmıştır.

```python
verisetim.sutun_sil('show_id')
```

## Yayın Yıllarının Dağılımı

Bu grafik, içeriklerin hangi yıllarda yoğunlaştığını göstermektedir.

```python
verisetim.dagilim_ciz("release_year")
```

## Temizlenmiş Verinin Kaydedilmesi

Uygulanan tüm veri temizleme işlemlerinden sonra elde edilen veri seti yeni bir CSV dosyasına kaydedilir.

```python
verisetim.veriyi_kaydet("veri/netflix_temiz.csv")
```
